# 01. Implementation and performance of Priority Queue

- Priority가 있는 element는 순서를 무시하고 선행됨

1. Lazy approach : 순서가 없는 실행
    - enqueue event : 단순 insert
    - dequeue event : Highest priority 를 찾아서 element를 remove
2. Early-bird approach : 순서가 있는 실행
    - enqueue event : 순서대로 insert
    - dequeue event : 순서 상 가장 첫 번째 것을 remove

[Implementation of priority queues]

- Sorted : Enqueue(node, priority)

- Enqueue = Insert
- Dequeue = Delete Highest Priority
- FindMax = Find Highest Priority

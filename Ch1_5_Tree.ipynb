# 03. Characteristics of Tree
[Tree의 특성]

- (Num of edge) = (Num of nodes) - 1
- Depth of Root = 0
- Height of root = height of tree
- Maximum num of nodes at level i with degree d = d^i
- Maximum num of leaves with height h and degree d = d^h
- Maximum size of tree with height h and degree d = 1 + d + d^2 + d^3 + ...+ d^h = (d^(h+1)-1)/(d-1)
- Height of a complete tree with size s and degree d = [logd(s(d-1)+1)의 올림]-1

# 05. Insert and Search Operation of Binary Search Tree

```python
# Insert Operation

def insert(self, value, node = ''):
    if node == '':
        node = self.root
    if self.root = TreeNode(value, ''):
        return
    if value == node.getValue():
        return
    if value > node.getValue():
        if node.getRHS() == '':
            node.setRHS(TreeNode(value, node))
        else:
            self.insert(value, node.getRHS())
    if value < node.getValue():
        if node.getLHS() == '':
            node.setLHS(TreeNode(value, node))
        else:
            self.insert(value, node.getLHS())
    return
```

```python
# Search Operation

def search(self, value, node=''):
    if node == '':
        node = self.root
    if value == node.getValue():
        return True
    if value > node.getValue():
        if node.getRHS() == '':
            return False
        else:
            return self.search(value, node.getRHS())
    if value < node.getValue():
        if node.getLHS() == '':
            return False
        else:
            return self.search(value, node.getLHS())
```

- RHS의 Minimum을 ROOT로 올리는 경우
    - RHS로 가서 LHS로 끝까지 내려가면 최솟값이 나옴

image.png

```python
# Maximum 찾기

def findMax(self, node = ''):
    if node == '':
        node = self.root
    if node.getRHS() == '':
        return node
    return self.findMax(node.getRHS())
```

```python
# Minimum 찾기

def findMin(self, node = ''):
    if node == '':
        node = self.root
    if node.getLHS() == '':
        return node
    return self.findMin(node.getLHS())    
```

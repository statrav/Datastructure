# 04. Quick Sort
[Quick Sort]

- Pivot 선택 (Pivot : 왼쪽과 오른쪽의 리스트를 분리하는 기준점 = threshold)
- 2 sequence로 나눔
- return : small scale로 recrusion
- 중간값을 잘 설정하는 것이 중요
- Merge sort는 항상 중간값으로 자르기 때문에, 항상 같은 sequence를 가짐 → Quick Sort는 sequence가 다를 수 있음

[importance of pivot in quick sort]

- O(N^2)의 sorting 알고리즘으로 구현
- Median / Random index로 pivot 선택
[implemention of qucik sort]

```python
import random

#Random Number Generation
N = 10
lstNumbers = list(range(N))
random.shuffle(lstNumbers)

print(lstNumbers)
```

```python
def performQuickSort(seq, pivot = 0):
    if len(seq) <= 1:
        return seq
    # Pivot selection : 항상 첫번째 element 선택
    pivotValue = seq[pivot]
    # 2개로 sequence 나누기
    less = []
    greater = []
    for itr in range(len(seq)):
        if itr == pivot:
            continue
        elif seq[itr] > pivotValue:
            greater.append(seq[itr])
        elif seq[itr] <= pivotValue:
            less.append(seq[itr])
    # Recursive calls
    ret = performQuickSort(less)+[pivotValue]+performQuickSort(greater)
    return ret

print(performQuickSort(lstNumbers))
```

# 05. Counting Sort

[Counting Sort]

- O(N) Sorting의 일환
- 가정1 : integer만 가능 (자유도 떨어짐)
- 가정2 : maximum value 는 K (0부터 K까지)
- comparison 과정 없이 occurance로 O(N) 만큼만 진행함

image.png

[Implementation of counting sort]

```python
# Random Number 생성

import random

N = 10
lstNumbers = list(range(N))
random.shuffle(lstNumbers)

print(lstNumbers)
```

```python
def performCountingSort(seq):
    max = -9999
    min = 9999
    # Counting 저장소 생성
    for itr in range(len(seq)):
        if seq[itr] > max:
            max = seq[itr]
        if seq[itr] < min:
            min = seq[itr]
    counting = list(range(max-min+1))
    for itr in range(len(counting)):
        counting[itr] = 0
    
    # Counting 준비
    for itr in range(len(seq)):
        value = seq[itr]
        counting[value-min] = counting[value-min]+1
    cnt = 0
    
    # Counted Number print
    for itr1 in range(max-min+1):
        for itr2 in range(counting[itr1]):
            seq[cnt] = itr1 + min
            cnt = cnt + 1
    return seq

print(performCountingSort(lstNumbers))
```

# 06. Radix Sort

[Radix Sort]

- 가정 : integer를 가진다
- 가장 큰 단위수까지 진행
- O(ND)

image.png

[Implementation of radix sort]

```python
# Random Number 생성

import random

N = 10
lstNumbers = list(range(N))
random.shuffle(lstNumbers)

print(lstNumbers)
```

```python
import math

def performRadixSort(seq):
    max = -999999
    
    # Digit Number 찾기
    for itr in range(len(seq)):
        if seq[itr] > max:
            max = seq[itr]
    D = int(math.log10(max))
    
    
    # Values를 buckets로 바꾸기
    for itr1 in range(0, D+1):
        buckets = []
        for itr2 in range(0, 10):
            buckets.append([])
        for itr2 in range(len(seq)):
            digit = int(seq[itr2] / math.pow(10, itr1)) %10
            buckets[digit].append(seq[itr2])
        
        # partially sorted된 values를 print
        cnt = 0
        for itr2 in range(0, 10):
            for itr3 in range(len(buckets[itr2])):
                seq[cnt] = buckets[itr2][itr3]
                cnt = cnt + 1
    return seq

print(performRadixSort(lstNumbers))
```

# 01. Hash

[Limit of Divide and Conquer]
- Divide and Conquer에 기반을 둔다면, 컴패리즌을 줄여줄 수는 있으나 없앨 수는 없음
- 아주 많은 데이터를 핸들링해야하는 경우에 여전히 한계가 존재하는 것

[Cheap storage cost and expensive time cost]
- 어떻게 '작업시간'을 줄일 것인가?
- Dictionary 형식으로 key-value pair로 저장한다면?
- Dictionary size는 최소한의 저장공간
- 그렇다면 Dictionary size 안에 있는 실제 index를 다시 만들어내서, 불필요한 key의 사용을 줄이는 function을 만들자 → Hash Table

# 02. Hash table

[Hash table]
- Hash table : key와 value가 저장되거 있는 매우 큰 사이즈의 Table
- Hash Fuction : key의 array index를 찾아주는 function
- 특성1 : 하나의 key가 하나의 index에 연결되어 있음
- 특성2 : 하나의 index에 여러가지의 ket를 가질 수는 있음 → 충돌
- Low cost, Determisitic, Variable range(Range에 딱맞는 결과가 나와야 함의 index의 형태를 가지는 function이 좋은 function

# 03. Examples of hash funtion

[Modulo based hash function]
- Modulo : nummeric ket가 input되었을 때 특정 Number로 나눠서, 나눈 값의 '나머지(%)'로 진행하는 방법
    - number는 hash table size로 지정
- size=10 일때, f(17)=7

[Square based hash function]
- Modulo는 key 정보를 변형하기 때문에 full 정보가 아님
- 어떻게 하면 full 정보를 다 쓸 수 있을까? → Square
- Mid square hash function :활용성이 적은 정보는 버림
    
    - N개를 선택할 때, 기준은 변동성이 있음 (N = mid-square digit)
    
[Digit analysis based hash function]
- 수 자체를 분석하는 것이 아닌, 수를 이루고 있는 '숫자'를 활용
- 먼저, 수를 이루는 숫자들을 각각 더해준다
- 그 다음, 해당 수로 Modulo based has funtion 진행
- ex. key:9603467 → 9+6+0+3+4+6+7=35 → f(35)

# 04. Collision resolution of hashing

[Collision resolution of hasing]
- Load factor (0과 1사이)
    - = N / S
        - N : Stored entries의 크기
        - S : Hash Table의 크기
    - 얼마나 한 index에 value가 붐비게 들어가 있는가
    
[Collision resolution by Closed addressing]
- 해당 index에 value를 그냥 다 넣음
- linked list화 됨
- 이 때에는 Load factor가 1 이상이 될 수도 있음 : 모든 index가 1개 이상의 entry를 가질 때

[Collision resolution by Open addressing]
- 한 index에 하나의 key만 들어감
- probing 과정 : entry가 이미 차있으면 다른 곳으로 이동해서 빈 entry로만 들어감
    - Linear probing : 바로 다음 entry로 이동 → f(key) + i
    - Quadratic probing : 점점 더 멀리 이동 → f(key) + i + i^2

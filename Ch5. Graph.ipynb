# 02. Representation of graph
[Data structure for graphs]
- Density = |E| / {|V|(|V|-1)}
- Array Representation : Adjacency Matrix
    - A[i][j] = matrix E 에 있으면 1 없으면 0
    
    - weighted matrix에서는 edge weight 값을 0과 1 대신 대입
    
    - 이 예시들은 0의 값이 많은 sparse한 matrix임 (메모리 공간의 활용성이 떨어짐)
    
[Adjacency List Representation for Sparse Graph]
- Sparse한 Graph를 표현할 때, Vertex로 시작하는 List의 방법을 활용 (효율적)
